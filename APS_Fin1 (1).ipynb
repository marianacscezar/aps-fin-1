### Ativos



```python
ativos= df.iloc[15, 2:7]
anos=[2019,2020,2021,2022,2023]
lista_ativos= ativos.tolist()
lista_ativos
```

```python
plt.bar(anos,lista_ativos, color= "#43627F")

# Adicionando títulos e rótulos aos eixos
plt.title('Evolução dos Ativos')
plt.xlabel('Anos')
plt.ylabel('Valor dos Ativos')
plt.grid()

# Exibindo o histograma
plt.show()
```
